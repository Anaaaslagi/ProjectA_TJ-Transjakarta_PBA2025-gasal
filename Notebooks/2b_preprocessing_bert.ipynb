```python
import pandas as pd
import re
import string

# 1. Load Data Hasil Scraping (Data Mentah)
df = pd.read_csv('../Data/hasil_scraping_transjakarta.csv', sep=';')
```

```python
# 2. Definisi Fungsi Preprocessing Khusus BERT (Minimal Cleaning)
def preprocess_for_bert(text):
    if not isinstance(text, str):
        return ""

    # a. Case Folding
    text = text.lower()

    # b. Menghapus URL
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)

    # c. Menghapus Mention (@user) dan Hashtag (#)
    text = re.sub(r'@\w+|#\w+', '', text)

    # d. Menghapus Angka
    text = re.sub(r'\d+', '', text)

    # e. Menghapus Tanda Baca
    text = text.translate(str.maketrans('', '', string.punctuation))

    # f. Menghapus Whitespace berlebih
    text = re.sub(r'\s+', ' ', text).strip()

    return text
```

```python
# 3. Terapkan Preprocessing
df['content_cleaned_bert'] = df['Isi Berita'].apply(preprocess_for_bert)

# 4. Simpan Hasil
# Kita simpan dengan nama berbeda agar tidak menimpa file lama
output_file = '../Data/hasil_preprocessing_bert.csv'
df.to_csv(output_file, index=False)

print(f"Preprocessing selesai! Data disimpan ke {output_file}")
print("Contoh data asli vs clean:")
print(df[['Isi Berita', 'content_cleaned_bert']].head())
```
